3.REINDEXING

Dataset

```python
data={'yillar':[2021,2020,2019,2018,2017,2016,2015,2010],
      'aholi soni': [33.9,33.5,32.9,32.5,31.9,31.4,30.9,28.5],
      'temp':[1.54,1.48,1.56,1.62,1.65,1.66,1.64,1.53],
      'zichlik':[79.77,78.68,77.53,76.34,75.13,73.9,72.71,67.03]}
df=pd.DataFrame(data)
df
```

Changing indexes

```python
df.index=[21,20,19,18,17,16,15,10]
df
```

Creating new indexes

```python
ind=[19,10,20,17,18,21,15,16]
ind.sort()
ind
```

Reindexing and changing first columns layouts

```python
df.reindex(ind)
```

```python
df
```

```python
df=df.reindex(ind)
df
```

```python
df.reindex([10,15,16])
```

Returning more indexes than actual dataset

```python
df.reindex(index=range(10,22))
```

Giving values to NaN elements

```python
df.reindex(index=range(10,22), method='nearest')
```

```python
df.reindex(index=range(10,22), method='ffill')
```

Dropping one of the columns

```python
df2=df.reindex(columns=['aholi soni', 'temp', 'zichlik'])
df2
```

to get one random sample

```python
df2.sample()
```

to get 3 random samples

```python
df2.sample(3)
```

returning a part of dataframe

```python
df.loc[[10,17,18,21],['aholi soni', 'temp']]
```
