Correlation in a pandas DataFrame refers to the statistical relationship between two or more columns within the DataFrame. It measures how much one column tends to move with or against another.

There are 2 ways to calculate correlation in pandas:

a) Using df.corr(): This method computes the pairwise correlation coefficients between all numerical columns in the DataFrame. It returns a new DataFrame with the correlation values.

b) Using df.corrwith(): This method calculates the correlation between a specific column and other columns (or another DataFrame).

(minus means reverse correlation)

```python
df.corr()
```

```python
df.corrwith(df['price'])
```

```python
df['price'].corr(df['mileage'])
```

```python
df['engineSize'].corr(df['mpg'])
```


In pandas DataFrames, the .value_counts() method is a powerful tool for analyzing the frequency of unique values within a specific column or across the entire DataFrame.

Functionality:

Counting unique values: It calculates the number of times each unique value appears in the specified column(s).
Output: It returns a pandas Series object, where the index represents the unique values and the values represent their corresponding counts.

Syntax:

Basic usage: df['column_name'].value_counts()
Counting in all columns: df.value_counts()

```python
df['model'].value_counts()
```

```python
df['fuelType'].value_counts()
```

```python
df['transmission'].value_counts()
```

Filtering

The .isin() method is a versatile tool for checking if values in a column exist within a specified list, Series, or other iterables. It returns a DataFrame of boolean values, indicating whether each element in the column matches any value in the reference set.

```python
models = [' B Class', ' CLK']
filtr = df['model'].isin(models)
filtr
```

```python
df[filtr]
```

```python
filtr = df['fuelType'].isin(['Diesel', 'Petrol'])
df[filtr]
```

```python
df[df.year>2018]
```

```python
df[(df.year>2018) & (df.price<20000)]
```

```python
df2 = df[(df.year>=2018) & (df.price<20000) & (df.mileage<10000)]
df2
```

```python
df[df.transmission == 'Automatic']
```
