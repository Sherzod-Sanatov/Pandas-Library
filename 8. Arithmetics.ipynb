Adding two Series

Series x and Series y

```python
s1=pd.Series([1.2,3.1,-4.2,-6.4,8.9], index=['a','b','d','f','g'])
s2=pd.Series([5.2,6.3,-4,-7.2,3.8,-0.54,0.67], index=['a','b','c','d','e','g','h'])
print(s1)
print(s2)
```

Index-wise adding takes place

The index that does not match has NaN value

For example, there is a index in both s1 and s2 Series, and in a index in s1+s2 Series, the value equals 1.2+5.2=6.4

And, there is f index in s1 Series, but there is not f index in s2 Series, so -6.4+NaN=NaN.

```python
s1+s2
```

Adding two DataFrame objects

```python
d1=pd.DataFrame(np.arange(9).reshape(3,3), columns=['a','b','c'], index=['Apple','Watermelon','Apricot'])
d2=pd.DataFrame(np.arange(12).reshape(4,3), columns=['a','c','d'], index=['Apple','Melon','Apricot','Pomegranate'])
print(d1)
print(d2)
```

Explanation for d1+d2:

1. The value is 2 in row=Apple and column=c in DataFrame d1

The value is 1 in row=Apple and column=c in DataFrame d2

After adding two values (d1+d2), the value becomes 3 in row=Apple and column=c

2. However, the value is 6 in row=Apricot and column=a in DataFrame d1

The value is 6 in row=Apricot and column=a in DataFrame d2

After adding two values (d1+d2), the value becomes 12 in row=Apricot and column=a


3. The value is 4 in row=Watermelon and column=b in DataFrame d1

The value is NaN in row=Watermelon and column=b in DataFrame d2

After adding two values (d1+d2), the value becomes NaN in row=Watermelon and column=a

or

The value is NaN in row=Watermelon and column=d in DataFrame d1

The value is NaN in row=Watermelon and column=d in DataFrame d2

So, after adding two values (d1+d2), the value is NaN in row=Watermelon and column=d

```python
d1+d2
```

There is add method to add two DataFrame

The advantage of this method over d1+d2 is that it is  possible to fill values with numbers like zero.

Explanation:

1. The value is 2 in row=Apple and column=c in DataFrame d1
The value is 1 in row=Apple and column=c in DataFrame d2

After adding two values (d1+d2), the value becomes 3 in row=Apple and column=c

2. However, the value is 6 in row=Apricot and column=a in DataFrame d1
The value is 6 in row=Apricot and column=a in DataFrame d2

After adding two values (d1+d2), the value becomes 12 in row=Apricot and column=a

or

The value is 4 in row=Watermelon and column=b in DataFrame d1
The value is NaN in row=Watermelon and column=b in DataFrame d2

After adding two values (d1+d2), the value becomes NaN in row=Watermelon and column=a

fill_value: the value is 1 in row=Apple and column=b in d1, the value is NaN in row=Apple and column=b, so the value after adding becomes 1 (not a NaN like above) because of fill_value method.

```python
d1.add(d2, fill_value=0)
```

Adding between Series and DataFrame

```python
df=pd.DataFrame(np.arange(12).reshape(3,4), index=list('abc'), columns=['apple', 'pomegranate', 'apricot', 'melon'])
sr=pd.Series([1,1,2,2], index=df.columns)
print(df)
print(sr)
```

column-wise adding takes place

for example: 0 (row=a, column=apple) + 1 (row=apple) = 1

4 (row=b, column=apple) + 1 (row=apple) = 5

8 (row=c,column=apple) + 1 (row=apple) = 9

and so on...


```python
df+sr
```
