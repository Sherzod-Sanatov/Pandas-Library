```python
import numpy as np
import pandas as pd
```

Dataset

```python
df = pd.DataFrame(np.random.randn(12).reshape(3,4), columns=['grape', 'apple', 'pomegranate', 'fig'], index=list('abc'))
df
```

absolute value

```python
np.abs(df)
```

rounded value

```python
np.round(df)
```

column-wise mean value

```python
df.mean()
```

index-wise mean value

```python
df.mean(axis=1)
```

Appling a function along a specified axis (rows or columns) of a DataFrame.

```python
df.apply(abs)
```

```python
df.apply(sum)
```

```python
df.apply(sum, axis=1)
```

```python
f=lambda x: x.max()-x.min()
df.apply(f)
```

```python
df.apply(lambda x: x.max()-x.min())
```

applying quite complex functions

```python
def foo(x):
  return pd.Series([x.max(), x.min()], index=['max', 'min'])

df.apply(foo)
```

Applying a function to every element (cell) of a DataFrame, element-wise.

Element-wise operation: Unlike apply() which works on rows or columns, applymap() applies the function to each individual cell.

```python
df=pd.DataFrame(np.arange(12).reshape(3,4), columns=['grape', 'apple', 'pomegranate', 'fig'], index=list('abc'))

l=lambda x: x*x

print(df)

print(df.applymap(l))
```
