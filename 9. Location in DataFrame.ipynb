```python
import pandas as pd
import numpy as np
```

```python
uzbekistan={20:33.5, 19:32.9, 18:32.5, 17:31.9, 16:31.4, 15:30.9}
kazakhistan={20:18.8, 19:18.5, 18:18.3, 17:18.1, 16:17.8, 15:17.6}
tajikistan={20:9.5, 19:9.3, 18:9.1, 17:8.9, 16:8.7, 15:8.5}
kirgizistan={20:6.5, 19:6.4, 18:6.3, 17:6.2, 16:6.1, 15:6.0}
turkmenistan={20:6.0, 19:5.9, 18:5.9, 17:5.8, 16:5.6, 15:5.6}

data={"Uzbekistan":uzbekistan,
      'Kazakhistan':kazakhistan,
      'Tajikistan':tajikistan,
      'Kyrgyzistan':kirgizistan,
      'Turkmenistan':turkmenistan}
df=pd.DataFrame(data)
df
```

loc stands for location

slicing rows of DataFrame using .loc method (attention on brakects[[]])

```python
df.loc[[20]]
```

```python
df.loc[[20,17]]
```

slicing rows and columns using .loc method

```python
df.loc[[18],['Kyrgyzistan']]
```

iloc stands for index location

slicing rows using .iloc[[]]

```python
df.iloc[[1]]
```

```python
df.iloc[[0,5]]
```

slicing rows and columns using .iloc

```python
df.iloc[[3],[2]]
```

extracting elements from a DataFrame using .at or .iat

```python
df.at[16, 'Uzbekistan']
```

iat stands for index at

```python
df.iat[4,0]
```

Using .at accessor: This method takes the label of the row and the column as arguments and allows direct assignment of a new value.

```python
import pandas as pd

data = {'Name': ['Alice', 'Bob', 'Charlie'], 'Age': [25, 30, 28]}
df = pd.DataFrame(data)
# Change the age of 'Bob' to 32
df.at['Bob', 'Age'] = 32
df
```
