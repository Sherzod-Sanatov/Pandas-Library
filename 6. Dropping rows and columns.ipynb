```python
import pandas as pd
```

```python
data={'yillar':[2021,2020,2019,2018,2017,2016,2015,2010],
      'aholi soni': [33.9,33.5,32.9,32.5,31.9,31.4,30.9,28.5],
      'temp':[1.54,1.48,1.56,1.62,1.65,1.66,1.64,1.53],
      'zichlik':[79.77,78.68,77.53,76.34,75.13,73.9,72.71,67.03]}
df=pd.DataFrame(data, index=[21,20,19,18,17,16,15,10])
ind=[19,10,20,17,18,21,15,16]
ind.sort()
df.reindex(ind)
```

to drop a row


last row is dropped

(default axis=0)

```python
df.drop(10)
```

to drop 2 or more than 2 rows

(default axis=0)

```python
df.drop([10,15])
```

to drop a column using axis

```python
df.drop('yillar', axis=1)
```

or

```python
df.drop('yillar', axis='columns')
```

in operations above, there would be no changes in original dataframe. So, to make changes in original dataframe, let's use methods below.

```python
df=df.drop(10)
df
```

or

```python
df.drop('yillar', axis=1, inplace=True)
df
```
